```python
# import threading, psutil, pynvml, time

# def monitor_resources(process, interval=0.9):
#     pynvml.nvmlInit()
#     handle = pynvml.nvmlDeviceGetHandleByIndex(0)
#     peak_cpu, peak_gpu = 0, 0
#     while getattr(threading.current_thread(), "running", True):
#         try:
#             cpu_mem = process.memory_info().rss / (1024 ** 2)
#             gpu_mem = pynvml.nvmlDeviceGetMemoryInfo(handle).used / (1024 ** 2)
#             peak_cpu = max(peak_cpu, cpu_mem)
#             peak_gpu = max(peak_gpu, gpu_mem)
#             print("cpu_mem:",cpu_mem)
#             print("gpu_mem:",gpu_mem)
        
#         except:
#             break
#         time.sleep(interval)
#     return peak_cpu, peak_gpu
import threading, time, psutil

try:
    import pynvml
    pynvml.nvmlInit()
    GPU_AVAILABLE = True
except:
    GPU_AVAILABLE = False

def monitor_resources(process, peaks, interval=0.2):
    """Monitor CPU + GPU memory usage in a background thread."""
    peaks["cpu"] = 0.0
    peaks["gpu"] = 0.0
    
    handle = None
    if GPU_AVAILABLE:
        handle = pynvml.nvmlDeviceGetHandleByIndex(0)
    
    while getattr(threading.current_thread(), "running", True):
        try:
            cpu_mem = process.memory_info().rss / (1024 ** 2)
            gpu_mem = 0.0
            if GPU_AVAILABLE:
                gpu_mem = pynvml.nvmlDeviceGetMemoryInfo(handle).used / (1024 ** 2)
            peaks["cpu"] = max(peaks["cpu"], cpu_mem)
            peaks["gpu"] = max(peaks["gpu"], gpu_mem)
        except psutil.NoSuchProcess:
            break
        time.sleep(interval)
```

```python
import time
import psutil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    classification_report, confusion_matrix, f1_score, accuracy_score, 
    precision_score, recall_score, roc_auc_score, log_loss, roc_curve
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier,AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelBinarizer
from sklearn.naive_bayes import GaussianNB
import json
import joblib
import os

# ======== MODEL LIST ========
models = {
    "LogisticRegression": LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=500),
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'NaiveBayes': GaussianNB(),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42),
    # 'KNN':KNeighborsClassifier(n_neighbors=5),
    # "GradientBoosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    # "SVM": SVC(probability=True)
}


results = []

# ======== CONTROLLED TRAINING ========
EPOCHS = 1  # For non-iterative models
STEPS_PER_EPOCH = 1



# ======== LOOP THROUGH MODELS ========
for model_name, model in models.items():
    cpu_mem_used=gpu_mem_used=elapsed_time=0
    process = psutil.Process(os.getpid())
    cpu_mem_before = process.memory_info().rss / (1024 ** 2)
    handle = None
    if GPU_AVAILABLE:
        handle = pynvml.nvmlDeviceGetHandleByIndex(0)
    if GPU_AVAILABLE:
        gpu_mem_before = pynvml.nvmlDeviceGetMemoryInfo(handle).used / (1024 ** 2)
    
    print(f"\nTraining model: {model_name}")
    process = psutil.Process(os.getpid())
    peaks = {}
    
    monitor_thread = threading.Thread(target=monitor_resources, args=(process, peaks))
    monitor_thread.running = True
    monitor_thread.start()
    
    start_time = time.time()
    model.fit(X_train, y_train)
    elapsed_time = time.time() - start_time
    
    # Stop thread safely
    monitor_thread.running = False
    monitor_thread.join()
    
    peak_cpu = peaks.get("cpu", 0.0)
    peak_gpu = peaks.get("gpu", 0.0)
    cpu_mem_used=peak_cpu-cpu_mem_before
    gpu_mem_used=peak_gpu-gpu_mem_before
    # # start_time = time.time()
    # # process = psutil.Process(os.getpid())
    # # mem_before = process.memory_info().rss / (1024 ** 2)  # in MB

    # # # Fit
    # # model.fit(X_train, y_train)
    # process = psutil.Process(os.getpid())
    # monitor_thread = threading.Thread(target=monitor_resources, args=(process,))
    # monitor_thread.running = True
    # monitor_thread.start()
    
    # start = time.time()
    # model.fit(X_train, y_train)
    # elapsed = time.time() - start
    
    # monitor_thread.running = False
    # monitor_thread.join()
    
    # peak_cpu, peak_gpu = monitor_resources(process)
    # # # Time & memory
    # # elapsed_time = time.time() - start_time
    # # mem_after = process.memory_info().rss / (1024 ** 2)  # in MB
    # # mem_used = mem_after - mem_before

    # # Predictions and probabilities
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test) if hasattr(model, "predict_proba") else None

    # ======= METRICS ========
    f1 = f1_score(y_test, y_pred, average='weighted')
    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred, average='weighted')
    precision = precision_score(y_test, y_pred, average='weighted')
    loss = log_loss(y_test, y_proba) if y_proba is not None else np.nan

    # Binarize y_test for AUC (multiclass)
    lb = LabelBinarizer()
    y_test_bin = lb.fit_transform(y_test)
    if y_proba is not None and y_test_bin.shape[1] == y_proba.shape[1]:
        auc = roc_auc_score(y_test_bin, y_proba, average='macro', multi_class='ovr')
    else:
        auc = np.nan

    cm = confusion_matrix(y_test, y_pred)
    # Macro & Weighted Avg (from classification_report)
    report = classification_report(y_test, y_pred, target_names=le.classes_, output_dict=True)

    # Per-class FPR, FNR
    per_class_metrics = {}
    for i, cls in enumerate(le.classes_):
        tp = cm[i, i]
        fp = cm[:, i].sum() - tp
        fn = cm[i, :].sum() - tp
        tn = cm.sum() - (tp + fp + fn)

        per_class_metrics[cls] = {
            "Precision": precision_score(y_test, y_pred, labels=[cls], average='macro', zero_division=0),
            "Recall": recall_score(y_test, y_pred, labels=[cls], average='macro', zero_division=0),
            "FNR": fn / (fn + tp + 1e-6),
            "FPR": fp / (fp + tn + 1e-6)
        }

    # print(report)
    # fp = cm.sum(axis=0) - np.diag(cm)
    # fn = cm.sum(axis=1) - np.diag(cm)
    # fpr = np.mean(fp / (fp + np.diag(cm) + 1e-6))  # Add epsilon to avoid div by zero
    # fnr = np.mean(fn / (fn + np.diag(cm) + 1e-6))
    tp_total = np.diag(cm).sum()
    fp_total = fp.sum()
    fn_total = fn.sum()
    tn_total = cm.sum() - (tp_total + fp_total + fn_total)
    
    global_fpr = fp_total / (fp_total + tn_total + 1e-6)
    global_fnr = fn_total / (fn_total + tp_total + 1e-6)

    
    # per_class_metrics = {}
    # for i, cls in enumerate(le.classes_):
    #     tp = cm[i, i]
    #     fp = cm[:, i].sum() - tp
    #     fn = cm[i, :].sum() - tp
    #     tn = cm.sum() - (tp + fp + fn)
    
    #     per_class_metrics[cls] = {
    #         "Precision": precision_score(y_test, y_pred, labels=[cls], average='macro', zero_division=0),
    #         "Recall": recall_score(y_test, y_pred, labels=[cls], average='macro', zero_division=0),
    #         "FNR": fn / (fn + tp + 1e-6),       # False Negative Rate
    #         "FPR": fp / (fp + tn + 1e-6)        # False Positive Rate
    #     }
    
    # # ===== Global FPR & FNR =====
    # tp_total = np.diag(cm).sum()
    # fp_total = cm.sum(axis=0) - np.diag(cm)   # sum of FP across classes
    # fn_total = cm.sum(axis=1) - np.diag(cm)   # sum of FN across classes
    
    # fp_total = fp_total.sum()
    # fn_total = fn_total.sum()
    
    # tn_total = cm.sum() - (tp_total + fp_total + fn_total)
    
    # global_fpr = fp_total / (fp_total + tn_total + 1e-6)
    # global_fnr = fn_total / (fn_total + tp_total + 1e-6)


    # Store summary results
    results.append({
        "Model": model_name,
        "F1_Score": f1,
        "Test_Accuracy": accuracy,
        "Test_Loss": loss,
        "Recall": recall,
        "Precision": precision,
        "AUC": auc,
        "CPU_Peak_MB": peak_cpu,
        "GPU_Peak_MB": peak_gpu,
        "Time_Seconds": elapsed_time,
        # "Memory_MB": mem_used
        "cpu_mem_used":cpu_mem_used,
        "gpu_mem_used":gpu_mem_used,
    })
    # results.append({
    #     "Model": model_name,
    #     "F1_Score": f1,
    #     "Accuracy": accuracy,
    #     "CPU_Peak_MB": peak_cpu,
    #     "GPU_Peak_MB": peak_gpu,
    #     "Train_Time": elapsed
    # })

    # Save detailed JSON report (per-class + macro/weighted)
    detailed_report = {
        "Model": model_name,
        "Overall": {
            "F1_Score": f1,
            "Test_Accuracy": accuracy,
            "Test_Loss": loss,
            "Recall": recall,
            "Precision": precision,
            "AUC": auc,
            "False_Positive_Rate": global_fpr,
            "False_Negative_Rate": global_fnr,
            "Time_Seconds": elapsed_time,
            # "Memory_MB": mem_used,
            "CPU_Peak_MB": peak_cpu,
            "GPU_Peak_MB": peak_gpu,
            "cpu_mem_used":cpu_mem_used,
            "gpu_mem_used":gpu_mem_used,
            
        },
        "Per_Class": per_class_metrics,
        "Classification_Report": report
    }

    with open(f"{model_name}_detailed_report.json", "w") as f:
        json.dump(detailed_report, f, indent=4)

    # Confusion Matrix Plot
    plt.figure(figsize=(12, 8))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues')
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.savefig(f"conf_matrix_{model_name}.png")
    plt.close()

# ======== SAVE METRICS CSV ========
results_df = pd.DataFrame(results)
results_df.to_csv("model_evaluation_results.csv", index=False)
print("\n✅ All models trained and evaluated. JSON reports + CSV saved.")

```
